```python
import requests
```

```python
url_1='https://www.google.com/search?'
p={'q':'Tim Berners-Lee'}
r1=requests.get(url_1,params=p)
print(r1.url)
```

```python
print(r1.status_code)
```

```python
print(r1.content)
```

```python
print(r1.headers)
```

```python
r2=requests.post('https://www.google.com',data = {'key':'value'})
```

```python
print(r2.content)
```

```python
print(r2.status_code)
```

```python
print(r2.headers)
```
